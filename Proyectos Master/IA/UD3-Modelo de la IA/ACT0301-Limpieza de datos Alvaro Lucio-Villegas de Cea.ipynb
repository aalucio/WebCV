```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mutual_info_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction import DictVectorizer
```

```python
df = pd.read_csv('../DataSets/WA_Fn-UseC_-Telco-Customer-Churn.csv')
print(len(df))
df.head().T
```

```python
df=df.drop_duplicates()
df.drop(columns="gender")
```

```python
total_charges = pd.to_numeric(df.TotalCharges, errors='coerce')
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
df.TotalCharges = df.TotalCharges.fillna(0)
df[total_charges.isnull()][['customerID', 'TotalCharges']]
```

```python
conver = lambda str: str.lower().str.replace(' ', '_').str.replace('-','_')

numerizar=lambda str: str.lower().str.replace('yes', '1').str.replace('no','0').str.replace('female','1').str.replace('male','0')


df.columns = conver(df.columns.str)


for col in list(df.dtypes[df.dtypes == 'object'].index):
    df[col] = conver(df[col].str)
    df[col] = numerizar(df[col].str)

df.head().T
```

```python
categorias = ['gender', 'seniorcitizen', 'partner', 'dependents','phoneservice', 'multiplelines', 'internetservice', 'onlinesecurity', 'onlinebackup', 'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies', 'contract', 'paperlessbilling', 'paymentmethod']

numericos=['tenure', 'monthlycharges', 'totalcharges']

#Cuenta las distan posibilidades que encuentra en una columna.
df[categorias].nunique()
```

```python
#Division de los datos

#En este caso valos a separar 80% para entramiento y un 20% para test
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)

df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=1)
y_train = df_train.churn.values
y_val = df_val.churn.values




del df_train['churn']
del df_val['churn']

df_train.head().T

df_train_full.head()
```

## Analisis de importancia


```python
global_mean = df_train_full.churn.mean()
round(global_mean, 3)
df_train_full.head()
```

```python

#Vamo a averiguar cuales columnas son mas importantes para el analisis para poder discriminarlas si es necesario.

media = df_train_full.churn.mean()
round(media, 3)

calculate_mi = lambda col: mutual_info_score(col, df_train_full.churn)

df_mi = df_train_full[categorias].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='Importancia')
df_mi
```

Podemos observar que las 3 ultimas columnas [gender,phoneservice,multiplelines] no son muy importantes para el analisis en cuestión.

## Entrenamiento del modelo y vectorización

```python
train_dict = df_train[categorias + numericos].to_dict(orient='records')
dict(sorted(train_dict[0].items()))
```

```python
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
X_train[0]
```

```python
dv.get_feature_names_out()
```

```python
model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorias + numericos].to_dict(orient='records')
X_val = dv.transform(val_dict)
y_pred = model.predict_proba(X_val)
y_pred
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
y_pred
```

```python
churn = y_pred >= 0.58
y_val=(y_val == '1').astype(int)
churn=(churn == True).astype(int)
```
